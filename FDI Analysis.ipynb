# Import Required Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("FDI data.csv")
```

# Load the Dataset

```python
df.head()
```

```python
df.set_index('Sector', inplace=True)
```

```python
df = df.transpose()
df.head()
```

# Overall statistics Info about dataset

```python
print('Descriptive Statistics: ')
df.describe()
```

# Total FDI Vs Years Bar Plot

```python
df['Total FDI'] = df.sum(axis = 1)
yearly_total = df['Total FDI']
plt.figure(figsize=(10,6))
yearly_total.plot(kind='bar', color='skyblue')
plt.title('Total FDI in India (2000-01 to 2016-17)')
plt.xlabel('Financial Year')
plt.ylabel('Total FDI (in million USD)')
plt.xticks(rotation = 45)
plt.show()
```

```python
df.head()
```

# Total FDI Vs Years Lineplot

```python
growth_rate = df.pct_change(axis = 1) * 100
plt.figure(figsize=(12, 8))
sns.lineplot(data=growth_rate, dashes=False)
plt.title("Sector-wise Annual Growth Rate (%)")
plt.xlabel("Financial Year")
plt.ylabel("Growth Rate (%)")
plt.xticks(rotation=45)
plt.legend(title='Sector', bbox_to_anchor=(1, 1), loc='upper left')
plt.show()
```

# Top 10 Best performing years

```python
sorted = df.sort_values(by='Total FDI', ascending=False)
sorted.head(10)
```

#Top 10 Sectors

```python
new_df = df.transpose()
new_df.drop(index='Total FDI', inplace=True)
new_df['Category Total'] = new_df.sum(axis = 1)
new_df.sort_values(by = 'Category Total', ascending=False, inplace = True)
new_df.head(10)
```

```python
top10_sectors = new_df.head(10)
top10_sectors = pd.DataFrame(top10_sectors['Category Total'])
```

```python
top10_sectors = top10_sectors.transpose()
```

# Top 10 Companies with highest FDI

```python
top10_sectors.rename(columns={'SERVICES SECTOR (Fin.,Banking,Insurance,Non Fin/Business,Outsourcing,R&D,Courier,Tech. Testing and Analysis, Other)': 'SERVICES SECTOR', 'CONSTRUCTION DEVELOPMENT: Townships, housing, built-up infrastructure and construction-development projects': 'CONSTRUCTION DEVELOPMENT'}, inplace = True)
plt.figure(figsize=(8,6))
sns.barplot(data = top10_sectors)
plt.xticks(rotation = 90)
plt.title('Top 10 Companies with highest FDI')
plt.ylabel('Total FDI')
plt.show()

```
